```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#https://stackoverflow.com/questions/53918187/python-requests-to-continue-your-browser-has-to-accept-cookies-and-has-to-hav
web_url = 'https://www.11v11.com/teams/manchester-united/tab/matches'
response = requests.get(web_url,
                        headers = {'User-Agent': 'Popular browser\'s user-agent'})
soup = BeautifulSoup(response.text, features= 'lxml')
```

```python
print(soup.title.text)
```

```python
season_link = soup.find_all(name= 'ul', attrs= {'class': 'linkList', 'id' : 'season'})
links = season_link[0].find_all('a')
links
```

```python
def get_all_season_link():
    '''
    '''
    link_area = soup.find_all(name= 'ul', attrs= {'class': 'linkList', 'id' : 'season'})
    seasons_link = link_area[0].find_all('a')
    seasons_link_list = []
    for link in seasons_link:
        seasons_link_list.append(link.get('href'))
    return seasons_link_list


seasons_link_list = get_all_season_link()
seasons_link_list
```

```python
def get_result_n_neareast_season(n):
    season_data = []
    for i in range(n):
        season_link = seasons_link_list[i]
        response = requests.get(url= season_link, headers = {'User-Agent': 'Popular browser\'s user-agent'})
        season_result = BeautifulSoup(response.text, 'lxml')
        tmp = season_result.find_all(name= 'table', class_= 'width580 sortable')
        all_match = tmp[0].find_all('td')
        
        
        
        for i in range(0, len(all_match) - 5, 5):
            match_data = []
            for j in range(5):
                match_data.append(all_match[i + j].text)
            # match_data.append(all_match[i + 4].text.strip('\n'))
            match_data[3] = match_data[3].strip('\n')
            season_data.append(match_data)
    return season_data

season_data = get_result_n_neareast_season(3)
        
```

```python
import pandas as pd
df = pd.DataFrame(season_data)
df.rename(columns={0: 'Date', 1: 'Match', 2: 'Result', 3: 'Score', 4: 'Competition'}, inplace=True)
```
